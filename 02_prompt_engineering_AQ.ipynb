# Prompt Engineering

The GPT models are, at their core, predictive models, meaning that they generate string of text by predicting the following words given previously provided and generated text. Thanks to the large parameter space and training sample that characterize both GPT-3.5 and GPT-4, they are quite capable at predicting text. These are the *base* models. However, for more specific applications, we might need certain instructions to be followed, like a certain output format, a specific structure, or a systematic description of steps to get to a final answer. We can enhace the utility of these very powerful base LLMs through a process called **Instruction-based tuning**, which consists in feeding the model an initial prompt that specifies the desired output. The process of designing adequate prompts for tuning the model is called **Prompt Engineering**.

````python
from openai import OpenAI
from getpass import getpass

api_key = getpass("Enter API key: ")
client = OpenAI(api_key = api_key)
````

We first build a simple helper function to process our queries. The chat completions endpoint requires a list of messages, each represented by a dictionary with two entries: the role of who sent the message and the content of the message. You will also notice an argument called `temperature`; this parameter can go from 0 to 2 and controls the randomness in the generation of text, with higher values being more random. A value of 0 does not completely get rid of randomness.

````python
def get_completion(prompt, model = "gpt-3.5-turbo-0125"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model = model,
        messages = messages,
        temperature = 0)
    return response.choices[0].message.content
````

## Chatbots

At their heart, the GPT models are predictive text engines. What makes them special is their capacity to find relationships between words within a text provided to them, generating text that takes those relationships into account, appending that to the provided text, and repeating this process until it finds a suitable ending point. What we have just described is, in very basic terms, the transformer architecture, which is the current paradigm for LLM science.

Through large transformers and lots of training data obtained from the internet, the GPT models are capable at predicting text such that they can be used for what is called *chat completion*. This means a chain of messages can be fed to the model, and it will be capable of completing the chain as if it were one of the interlocutors. This approach is very popular, and the examples we have seen use this approach: our prompts are fed into the model as if they came from a person with the role `"user"`, and the model generates text taking the role of `"assistant"`. The list of dictionaries that the Chat Completions endpoint uses is the way we represent this chain of messages, so that the model can process it. This has the advantage of being able to add generated answers to the chain, as well as new queries. This way, the assistant has context for the rest of the conversation; we call this **memory**. Memory makes it so that the model identify references to preceding queries and responses; in linguistic terms, it understands [anaphora](https://en.wikipedia.org/wiki/Anaphora_(linguistics)) outside of the immediate query.

We will now set up an example of how this pans out. First, we set up a function that takes a message and processes it to have the format necessary for sending it to the API's endpoint, and then gets the response.

````python
def getCompletion(prompt, model = "gpt-3.5-turbo-0125"):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model = model,
        messages = messages,
        temperature = 0)
    return response.choices[0].message.content
````

## Self-Consistency

One of the more advanced techniques out there for prompt engineering is self-consistency. Proposed by [Wang et al. (2022)](https://arxiv.org/abs/2203.11171), self-consistency aims "to replace the naive greedy decoding used in chain-of-thought prompting". The idea is to sample multiple, diverse reasoning paths through few-shot CoT, and use the generations to select the most consistent answer. This helps to boost the performance of CoT prompting on tasks involving arithmetic and commonsense reasoning.

````python
def get_completion(query, system_prompt, message_history, model = "gpt-4o-mini", temperature = 1):
    messages = [{"role": "system", "content": system_prompt}]
    messages += message_history
    messages += [{"role": "user", "content": query}]
    response = client.chat.completions.create(
        model = model,
        messages = messages,
        temperature = temperature)
    return response.choices[0].message.content
````

Self-consistency involves generating several answers with possibly different reasoning paths, each with corresponding results and then aggregating them into one final answer. The aggregation can be done by majority vote (specially in the case of numerical questions), meaning that the final answer is a reasoning path that leads to the most common result.

````python
import json

class SelfConsistencyAssistant:
    def __init__(self, model = "gpt-4o-mini", temperature = .8, n_paths = 3):
        self.model = model
        self.temperature = temperature
        self.n_paths = n_paths
        self.examples = sc_examples
        self.aggregation_prompt = """You will be provided with several different answers to one question, and each answer may also contain a reasoning
        In a JSON list, include the answer's index, as ints, of the most common answer.
        """
        self.answer_prompt = "Your are a helpful assistant. When asked for an answer, detail your reasoning behind it"
        self.answer_template = "Answer {index}:\n{answer}\n\n"
        self.reason_path_history = [{"role": "system", "content": self.answer_prompt}] + self.examples
        self.final_prompt = "Answer the question using. Your answer must sumarize the reasoning of the other answers provided"
        self.aggregation_template = """Answers: {answers}
        Question: {question}"""

    def get_completion(self, messages):
        response = client.chat.completions.create(
            model = self.model,
            messages = messages,
            temperature = self.temperature)
        return response.choices[0].message.content

    def get_answer(self, query):
        messages = self.reason_path_history + [{"role": "user", "content": query}]
        answer = self.get_completion(messages)
        return answer

    def get_all_answers(self, query):
        self.answers = []
        self.parsed_answers = ""

        for i in range(self.n_paths):
            answer = self.get_answer(query)
            parsed_answer = self.answer_template.format(index=i, answer=answer)
            self.answers.append(answer)
            self.parsed_answers += parsed_answer

    def aggregate_answers(self):
        messages = [{"role": "system", "content": self.aggregation_prompt}]
        messages.append({"role": "user", "content": self.parsed_answers})
        aggregate_response = self.get_completion(messages).strip("`").strip("json").strip("\n")
        self.aggregate_response = aggregate_response
        consensus_answers_index = json.loads(aggregate_response)
        self.consensus_answers = [self.answers[i] for i in consensus_answers_index]

    def generate(self, query):
        self.get_all_answers(query)
        self.aggregate_answers()
        messages = [
            {"role": "system", "content": self.final_prompt},
            {"role": "user", "content": self.aggregation_template.format(answers=self.consensus_answers, question=query)}
        ]
        return self.get_completion(messages)

````

### Using knowledge generation

````python
class GeneratedKnowledgeAssistant:
    def __init__(self, model = "gpt-4o-mini"):
        self.model = model
        self.gen_knowledge_prompt = "Generate some knowledge relating to the input's topic without directly answering"
        self.gen_knowledge_examples = [
            {"role": "user", "content": "Input: Greece is larger than mexico."},
            {"role": "assistant", "content": "Knowledge: Greece is approximately 131,957 sq km, while Mexico is approximately 1,964,375 sq km, making Mexico 1,389% larger than Greece."},
            {"role": "user", "content": "Input: Glasses always fog up."},
            {"role": "assistant", "content": "Condensation occurs on eyeglass lenses when water vapor from your sweat, breath, and ambient humidity lands on a cold surface, cools, and then changes into tiny drops of liquid, forming a film that you see as fog. Your lenses will be relatively cool compared to your breath, especially when the outside air is cold."},
            {"role": "user", "content": "Input: A fish is capable of thinking."},
            {"role": "assistant", "content": "Fish are more intelligent than they appear. In many areas, such as memory, their cognitive powers match or exceed those of ’higher’ vertebrates including non-human primates. Fish’s long-term memories help them keep track of complex social relationships."},
            {"role": "user", "content": "Input: A rock is the same size as a pebble."},
            {"role": "assistant", "content": "A pebble is a clast of rock with a particle size of 4 to 64 millimetres based on the Udden-Wentworth scale of sedimentology. Pebbles are generally considered larger than granules (2 to 4 millimetres diameter) and smaller than cobbles (64 to 256 millimetres diameter)."}
        ]
        self.gen_answer_prompt = "Use the knowledge provided to answer the question"
        self.query_template = """Question: {query}
        Knowledge: {knowledge}"""

    def generate_knowledge(self, query):
        messages = [{"role": "system", "content": self.gen_knowledge_prompt}] + self.gen_knowledge_examples
        messages += [{"role": "user", "content": query}]
        response = client.chat.completions.create(
            model = self.model,
            messages = messages,
            temperature = self.temperature
        )
        return response.choices[0].message.content

    def generate(self, query):
        knowledge = self.generate_knowledge(query)
        messages = [{"role": "system", "content": self.gen_answer_prompt}]
        messages += [{"role": "user", "content": self.query_template.format(query=query, knowledge=knowledge)}]
        response = client.chat.completions.create(
            model = self.model,
            messages = messages,
            temperature = self.temperature
        )
        return response.choices[0].message.content, knowledge
````

````python
knowledge_assistant = GeneratedKnowledgeAssistant()
````

````python
answer, knowledge = knowledge_assistant.generate("Part of golf is trying to get a higher point total than others. Yes or No?")
print(answer)
````

## Tree-of-Thoughts

For complex tasks that require exploration or strategic lookahead, traditional or simple prompting techniques fall short. [Yao et al. (2023)](https://arxiv.org/abs/2305.10601) and [Long (2023)](https://arxiv.org/abs/2305.08291) proposed Tree of Thoughts (ToT), a framework that generalizes over chain-of-thought prompting and encourages exploration over thoughts that serve as intermediate steps for general problem solving with language models.





A simplified prompting technique proposed by [Hulbert (2023)](https://github.com/dave1010/tree-of-thought-prompting) relies on just asking the LLM to frame its response as a discussion between experts. This renders similar performance, although it does not have a way to check for the validity of intermediate steps with a state evaluator as in the other, more more formal approaches

````python
class SimpleToT:
    def __init__(self, model = "gpt-4o-mini", temperature = .8):
        self.model = model
        self.temperature = temperature
        self.prompt = """Imagine three different experts are answering a user's question.
        All experts will write down 1 step of their thinking,
        then share it with the group.
        Then all experts will go on to the next step, etc.
        If any expert realises they're wrong at any point then they leave."""
        self.query_template = "Question: {question}"

    def generate(self, query):
        messages = [{"role": "system", "content": self.prompt}]
        messages += [{"role": "user", "content": self.query_template.format(question=query)}]
        response = client.chat.completions.create(
            model = self.model,
            messages = messages,
            temperature = self.temperature)
        return response.choices[0].message.content
````

````python
simple_tot = SimpleToT()
````

````python
query = """Bob is in the living room.
He walks to the kitchen, carrying a cup.
He puts a ball in the cup and carries the cup to the bedroom.
He turns the cup upside down, then walks to the garden.
He puts the cup down in the garden, then walks to the garage.
Where is the ball?"""

print(simple_tot.generate(query))
````
